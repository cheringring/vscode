# 집계함수 

sum(): 모든 요소의 합
min(): 모든 요소중 최소값
max(): 모든 요소중 최대값
argmax(): 모든 요소 중 최대값의 인덱스
cumsum(): 모든 요소의 누적합



```python
# 0부터 7까지 제곱된 값을 2,4크기로 출력




a= np.arange(8).reshape(2,4)**2

print(a)

print()

print(a.sum())
print(a.min())
print(a.max())
print(a.argmax())
print(a.cumsum())

```

```python
arr = np.arange(9).reshape(3,3)

print(arr)
print()


print(arr.cumsum(axis=0)) #exis= 축 = 세로 방향으로 누적 

print()

print(arr.cumsum(axis=1)) # exis= 축 = 가로 방향으로 누적됨.

# 2차원배열이면 세로 먼저 0행 , 가로 1행 이니까. 방향 그리면서 이해하면 됨.

```

```python
sns.scatterplot(x='age',y='fare',hue='survived',data= titanic)


sns.regplot(x='age',y='fare',data= titanic[titanic['survived']== 0]
            ,scatter=False,color='red',label='Not survived')
sns.regplot(x='age',y='fare',data= titanic[titanic['survived']== 1]
            ,scatter=False,color='blue',label='survived')

# 데이터 타이타닉 조건 설바이브 = 0 과 설바이브 =1 둘다 리그레이션 그래프를 보여준다.

plt.legend() # 리그레이션 함수에 옆에 네모박스에 있는 글을 추가해줌.

plt.show()
```

```python
# 나이와 생존에 대한 시각화


sns.kdeplot(x= 'age', data=titanic,color='blue',label='age') #, shade=True
sns.kdeplot(x= 'age',data=titanic[titanic['survived']==1],color='red',label='survived')

plt.xlabel = 'age'
plt.legend()
plt.show()
```

```python
# 성별 기준으로 나이별 생존 시각화

# 자르기 ( hue는 색상으로 구분 )

#g= sns.kdeplot(data=titanic['sex'],data= titanic['survived'])
g = sns.FacetGrid(titanic,col='sex',hue='survived')

g.map(sns.kdeplot,'age',fill=True)


plt.legend()
# shade= fill 
plt.show()
```

```python
g = sns.FacetGrid(titanic,row='pclass',col='sex',hue='survived')
g.map(sns.kdeplot,'age',fill=True)


# row 는 pclass 별로 출력하라 라는 뜻 .



g.add_legend()
plt.show()
```
