```python

def insta_search(word):
    url = 'https://www.instagram.com/explore/tags/'+word
    return url



def select_first(driver):
    post = driver.find_elements(By.CSS_SELECTOR,'div > ._aabd._aa8k._al3l')[0]
    post.click()
    time.sleep(2)


def get_contents(driver):
    html = driver.page_source
    soup= BeautifulSoup(html,'html.parser')
    contents = soup.select('div> ._a9z6._a9za')
    contents_ = contents[0].text

    tags = hash_list=contents_.split('#') # #ê¸°ì¤€ìœ¼ë¡œ ì˜ë¼ì„œ
                                 # tags ì— ì§‘ì–´ë„£ê³ 
    content_data = tags[0]

# tags 1 ë¶€í„° ëê¹Œì§€ tagsì— ë‹¤ì‹œ ì§‘ì–´ë„£ê³  ê³µë°±ì œê±°.
    tags=tags[1:]
    tags=[item.strip() for item in tags]

    data = [content_data,tags]
    return data     

```

```python
def move_next(driver):


    try:
        next_button = driver.find_elements(By.CSS_SELECTOR,'div._aaqg._aqqh > button._abl-')[0]
        next_button.click()
        time.sleep(3)
        return True
    except: 
        return False
```

```python

word = 'ì œì£¼ë„ë§›ì§‘'
url = insta_search(word)
driver.get(url)
time.sleep(5)

select_first(driver)


# ìˆ˜ì§‘í•  ê²Œì‹œë¬¼ ìˆ˜ë¥¼ 50ê°œ ì •ë„ ê³ ì • 

results = []

for i in range(50):
    data = get_contents(driver) # get_contents í•¨ìˆ˜ê°€ ë°ì´í„°ë¥¼ ë¦¬í„´
    results.append(data)
    is_next = move_next(driver)
    if is_next == False:
        break


results
    

df = pd.DataFrame(results,columns = ['content','tags'])
df.to_excel('data/crawing_jeju.xlsx',index=False)
```
