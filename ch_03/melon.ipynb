```python
# selenium 으로 활용
                 # css ~ 그냥 쓰고, 뒤에 tbody는 하위
songs = driver.find_elements('css selector','tbody > tr')
len(songs) # variable) songs: List[WebElement]
```

```python
                

songs = driver.find_elements('css selector','tbody > tr')

for song in songs:
    title = song.find_elements('css selector','div.ellipsis.rank01 > span > a')[0].text
    singer = song.find_elements('css selector','div.ellipsis.rank02 > a')[0].text

    print( title,' | ', singer)

    
```

```python
# 위에 songs가 리스트타입으로 지정되어 있어 driver대신 songs를 쓰면 Sub~오류가남.



title = driver.find_elements('css selector','div.ellipsis.rank01 > span > a')
singer = driver.find_elements('css selector','div.ellipsis.rank02 > a')




for title,singer in zip(title,singer):  #랭(숫자)가 다 같아야함.
    # 하나의 튜플로 묶임 그래서 [0]이 필요가 없다. 
    #>>> list(zip('abcdefg', range(3), range(4)))
     #[('a', 0, 0), ('b', 1, 1), ('c', 2, 2)]
    # 이런식으로 묶임. zip 설명참조.

    print(title.text,'|',singer.text)


# [0]을 썼는데 오류가 떴다.
    # 하나의 튜플로 묶임 그래서 [0]이 필요가 없다.   < 이유다.


    
```

```python

import time
from selenium.webdriver.common.keys import Keys

url = 'https://www.melon.com/chart/index.htm'
driver = webdriver.Chrome()
driver.get(url)

# 얘가 로드할 시간을 충분히 준다 2초
time.sleep(2)

search_query = 'Dynamite'

search_Box = driver.find_elements('css selector','#top_search')

search_Box[0].clear()
search_Box[0].send_keys(search_query)
search_Box[0].click()
search_Box[0].send_keys(Keys.RETURN)


time.sleep(2)


from bs4 import BeautifulSoup
html = driver.page_source
soup= BeautifulSoup(html,'html.parser')


title = soup.find_elements('a.fc_gray')
singer = soup.find_elements('div#artistName > a.fc_mgray')


for title, singer in zip(title,singer):
    print(title.text , ' | ' , singer.text)
```
